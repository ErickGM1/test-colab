```python
#!/usr/bin/env python3
import gzip
import shutil
import csv
from pathlib import Path
from datetime import datetime

filename_csv = '/content/drive/My Drive/procesador_de_archivos/database01.csv'
# Define directories
zips_folder = Path('/content/drive/My Drive/procesador_de_archivos/zips')
files_folder = Path('/content/drive/My Drive/procesador_de_archivos/files')

# # Create files folder if it doesn't exist
# zips_folder.mkdir(exist_ok=True)
# files_folder.mkdir(exist_ok=True)

# ### Dev only ########################################################
# originals_folder = Path("originals")

# # Loop through all files in originals folder
# for file in originals_folder.glob("*.gz"):
#     if file.is_file():
#         destination = zips_folder / file.name
#         print(f"Copying {file.name} to zips folder")
#         shutil.copy2(file, destination)
#         print(f"  ✓ Successfully copied {file.name}")

# print("\nAll files copied successfully!")
# ### Dev only ########################################################

# a) Create a dictionary to hold the column definition
column_definitions_r0 = {
    'tipo_de_registro_r0': {'start': 1-1, 'end': 1},
    'empresa_r0': {'start': 2-1, 'end': 7},
    'fecha_r0': {'start': 8-1, 'end': 15},
    'cuenta_de_cheques_r0': {'start': 16-1, 'end': 31},
    'filler_r0': {'start': 32-1, 'end': 250},
}

print("Column Definitions r0:")
for col, defs in column_definitions_r0.items():
    print(f"  {col}: Start={defs['start']}, End={defs['end']}")

# a) Create a dictionary to hold the column definition
column_definitions_r2 = {
    'tipo_de_registro': {'start': 1-1, 'end': 1},
    'folio': {'start': 2-1, 'end': 22},
    'medio_de_pago': {'start': 23-1, 'end': 24},
    'forma_de_pago': {'start': 25-1, 'end': 26},
    'importe_bruto': {'start': 27-1, 'end': 40},
    'importe_descuento_recargo': {'start': 41-1, 'end': 48},
    'identificador_descuento_recargo': {'start': 49-1, 'end': 49},
    'importe_neto': {'start': 50-1, 'end': 63},
    'referencia_1': {'start': 64-1, 'end': 83},
    'referencia_2': {'start': 84-1, 'end': 103},
    'referencia_3': {'start': 104-1, 'end': 123},
    'referencia_4': {'start': 124-1, 'end': 143},
    'fecha_de_vencimiento': {'start': 144-1, 'end': 151},
    'sucursal': {'start': 152-1, 'end': 155},
    'hora': {'start': 156-1, 'end': 161},
    'estatus': {'start': 162-1, 'end': 162},
    'codigo_de_rechazo': {'start': 163-1, 'end': 164},
    'tipo_de_movimiento': {'start': 165-1, 'end': 165},
    'fecha_de_pago': {'start': 166-1, 'end': 173},
    'descripcion_forma_de_pago': {'start': 174-1, 'end': 179},
    'clave_de_banco': {'start': 180-1, 'end': 182},
    'numero_de_cheque': {'start': 183-1, 'end': 189},
    'filler': {'start': 190-1, 'end': 244},
    'numero_de_empresa': {'start': 245-1, 'end': 250},
}

print("Column Definitions r2:")
for col, defs in column_definitions_r2.items():
    print(f"  {col}: Start={defs['start']}, End={defs['end']}")

# a) Create a dictionary to hold the column definition
column_definitions_r4 = {
    'tipo_de_registro_r4': {'start': 1-1, 'end': 1},
    'numero_pagos_acreditados_r4': {'start': 2-1, 'end': 7},
    'importe_pagos_acreditados_r4': {'start': 8-1, 'end': 21},
    'numero_pagos_devueltos_r4': {'start': 22-1, 'end': 27},
    'importe_pagos_devueltos_r4': {'start': 28-1, 'end': 41},
    'numero_pagos_cancelados_r4': {'start': 42-1, 'end': 47},
    'importe_pagos_cancelados_r4': {'start': 48-1, 'end': 61},
    'numero_total_pagos_r4': {'start': 62-1, 'end': 67},
    'importe_total_pagos_r4': {'start': 68-1, 'end': 81},
    'indicador_r4': {'start': 82-1, 'end': 250},
}

print("Column Definitions r4:")
for col, defs in column_definitions_r4.items():
    print(f"  {col}: Start={defs['start']}, End={defs['end']}")

# Loop through all .gz files in zips folder
for gz_file in zips_folder.glob("*.gz"):
    # Get output filename (remove .gz extension)
    output_file = files_folder / gz_file.stem

    print(f"Unzipping {gz_file.name} to {output_file.name}")

    # Decompress the file
    with gzip.open(gz_file, 'rb') as f_in:
        with open(output_file, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)

    print(f"  ✓ Successfully unzipped to {output_file}")

print("\nAll files unzipped successfully!")

### Util
def parse_file_to_dict(file_path):
    line_dict = {}

    with open(file_path, 'r') as file:
        for idx, line in enumerate(file, start=1):
            if len(line.strip()) > 0:  # Only consider non-empty lines
                raw_line = line.rstrip('\n')
                line_dict[idx] = raw_line  # Key = line number, Value = line content

        # Print the dictionary
        print("\n--- Dictionary of Lines ---")
        for key, value in line_dict.items():
            print(f"{key}: {value}: {len(value)}")

        return line_dict

def parse_fixed_width_line(line, definitions):
    parsed_data = {}
    for column_name, pos in definitions.items():
        start = pos['start']
        end = pos['end']

        column_value = line[start:end].strip()
        if column_name in ['importe_bruto', 'importe_descuento_recargo', 'importe_neto']:
            try:
                # Assuming the value is an integer representing cents
                column_value = f"${float(column_value) / 100:.2f}"
            except ValueError:
                column_value = column_value # Keep as is if conversion fails
        elif column_name == 'referencia_1':
            column_value = column_value.lstrip('0')
        elif column_name in ['fecha_de_vencimiento', 'fecha_de_pago']:
            try:
                # Parse YYYYMMDD and format to YYYY-MM-DD
                column_value = datetime.strptime(column_value, '%Y%m%d').strftime('%Y-%m-%d')
            except ValueError:
                column_value = column_value # Keep as is if conversion fails
        elif column_name == 'hora':
            try:
                # Parse HHMMSS and format to HH:MM:SS AM/PM
                column_value = datetime.strptime(column_value, '%H%M%S').strftime('%I:%M:%S %p')
            except ValueError:
                column_value = column_value # Keep as is if conversion fails

        print(f"*** Column Name: {column_name} - Column value: {column_value}")
        parsed_data[column_name] = column_value

    return parsed_data

def parse_dictionary(line_dict, file_name):
    all_items = []
    line_parsed_dict = []

    # Split file name (e.g., "file.1" -> ["file", "1"])
    file_name_parts = file_name.split('.')
    file_name_dict = {
        'archivo_ra': file_name,
        'rccg_tipo_archivo_ra': file_name_parts[0],
        'fecha_generacion_archivo_ra': file_name_parts[1],
        'numero_empresa_ra': file_name_parts[2]
    }

    for line_number, line_content in line_dict.items():
        if line_number == 1:
            print("\nlines 1:")
            parsed_line_header = parse_fixed_width_line(line_content, column_definitions_r0)
        elif line_number == len(line_dict):
            print("\nlines 3:")
            parsed_line_footer = parse_fixed_width_line(line_content, column_definitions_r4)
        else:
            print("\nlines 2:")
            parsed_line_detail = parse_fixed_width_line(line_content, column_definitions_r2)

            line_parsed_dict.append(parsed_line_detail)

    for item in line_parsed_dict:
        merged_dict = {**item}
        all_items.append(merged_dict)
        print("\ntwosss lines:", item)
    print("\nones lines:", parsed_line_header)
    print("\n3sssss lines:", parsed_line_footer)
    return all_items

def test_csv_write(all_items):
    # Check if file exists to determine if we need to write headers
    file_exists = Path(filename_csv).exists()

    # Read existing rows if file exists
    existing_rows = set()
    if file_exists:
        with open(filename_csv, mode="r", newline="") as file:
            reader = csv.reader(file)
            next(reader, None)  # Skip header
            for row in reader:
                # Create tuple of row values to store in set for comparison
                existing_rows.add(tuple(row))

    # Open file in append mode
    with open(filename_csv, mode="a", newline="") as file:
        writer = csv.writer(file)

        # Write header (keys) only if file doesn't exist
        if not file_exists:
            writer.writerow(all_items[0].keys())

        # Write values (append only if not exists)
        new_rows = 0
        skipped_rows = 0
        for item in all_items:
            row_values = tuple(item.values())
            if row_values not in existing_rows:
                writer.writerow(item.values())
                existing_rows.add(row_values)  # Add to set to avoid duplicates within same batch
                new_rows += 1
            else:
                skipped_rows += 1

    print(f"Data {'appended to' if file_exists else 'written to'} {filename_csv} successfully. New rows: {new_rows}, Skipped duplicates: {skipped_rows}")


# Loop through all files in files folder and parse each one
for file in files_folder.iterdir():
    if file.is_file():
        print(f"\n{'='*60}")
        print(f"Processing file: {file.name}")
        print(f"{'='*60}")
        files = parse_file_to_dict(file)
        all_items = parse_dictionary(files, file.name)
        test_csv_write(all_items)
        print("\nAll items:", all_items)
print("\n\nAll files processed successfully!")
```
